```python
import numpy as np
from mpmath import findroot
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
```

```python
def f1(x,y):
    f1 = 2*x**2+y**2-2
    return f1

def f2(x,y):
    f2 = (x-0.5)**2+(y-1)**2-0.25
    return f2
```

```python
x = np.linspace(-50, 50, 200)
y = np.linspace(-50, 50, 200)
X, Y = np.meshgrid(x, y)

Z1 = f1(X,Y)
Z2 = f2(X,Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z1, 1000, cmap='viridis')
ax.contour3D(X, Y, Z2, 1000, cmap='plasma')
ax.view_init(35, 60)
```

```python
root1 = findroot([f1,f2],(0+0j,1+0j))
root2 = findroot([f1,f2],(1+0j,0+0j))
root3 = findroot([f1,f2],(-1+1j,2+1j))
root4 = findroot([f1,f2],(-1-1j,2-1j))
print(root1, "\n")
print(root2, "\n")
print(root3, "\n")
print(root4, "\n")
```
