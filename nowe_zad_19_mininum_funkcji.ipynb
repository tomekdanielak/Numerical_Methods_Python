```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import scipy.optimize as optimize
import random
%matplotlib inline
```

```python
def f(x, y):
    return 0.25*x**3 - 0.5*x**2 + 0.125*x + 0.0625*x + y**2 - 0.0625*y

x = np.linspace(-50, 50, 200)
y = np.linspace(-50, 50, 200)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 1000)
ax.view_init(35, 60)
```

```python
def f(params):
    x, y = params
    return 0.25*x**3 - 0.5*x**2 + 0.125*x + 0.0625*x + y**2 - 0.0625*y


for i in range(1,40):
    x = random.uniform(-3,3)
    y = random.uniform(-3,3)
    
    initial_guess = [x,y]
    result = optimize.minimize(f, initial_guess, tol=0.00001)
    print(result.x)
    

""" Dla przejrzystoci minima zostay obliczone tylko z 40 punktw pocztkowych, a nie ze 128 jak w treci zadania. 
Wida e cz z nich si powtarza (w granicach zadanej tolerancji) - tzn. e z rncyh punktw startowych funkcja zbiega si do punktw staych"""
```
