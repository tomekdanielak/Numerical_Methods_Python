```python
import numpy as np


""" Przykład A """

A = np.zeros((7,7),dtype=float)   # definiuje macierz A
np.fill_diagonal(A,4)

for i in range(6):
    A[i][i+1] = 1
    A[i+1][i] = 1
A = np.asarray(A)
A = np.float32(A)
    
vector_b = np.asarray([i for i in range(1,8)])
vector_b = np.float32(vector_b)

    
def cholesky_decomposition(matrix):
    """Rozkład Cholesky'ego uwzględniający strukture macierzy rzadkiej"""
    
    n = len(matrix)
    
    C = [[0.0] * n for i in range(n)]  # Tworze macierz zerową, o takim samym rozmiarze jak A
    
    for i in range(n):
        for k in [k for k in range(i-1,i+1) if k > -1]: # wybieram tylko te elementy które nie są zerowe
            tmp_sum = sum(C[i][j] * C[k][j] for j in range(k))
            
            if (i == k):
                C[i][k]  = np.sqrt(matrix[i][i] - tmp_sum)
            else:
                C[i][k] = (1.0 / C[k][k] * (matrix[i][k] - tmp_sum))
    C = np.asarray(C)
    return C
"""
Nasze równanie to teraz CC.Tx = b
Rozwiązuje C * C.T * x = b
Na początku C * y = b za pomocą forward substitution gdzie y = C.T * x
Następnie za pomocą backward substitution rozwiązuje C.T * x = y
"""


def forward_substitution(matrix_A, vector_b): 
    """Ax = b - ogólny zapis równania
    Algorytm uwzględnia strukture macierzy rzadkiej"""
    
    vector_x = np.zeros_like(vector_b)
    vector_x = np.float32(vector_x)
    
    n = len(matrix_A)
    
    for i in range(n):
        if i == 0:
            vector_x[i] = vector_b[i] / matrix_A[i][i]
        else:
            vector_x[i] = (vector_b[i] - matrix_A[i][i-1]*vector_x[i-1]) / matrix_A[i][i]
            
    return vector_x


def backsubstitution(matrix_A, vector_b):
    """Ax = b - ogólny zapis równania
    Algorytm uwzględnia strukture macierzy rzadkiej"""
        
    vector_x = np.zeros_like(vector_b)
    vector_x = np.float32(vector_x)
    
    n = len(matrix_A)

    for i in range(n-1, 0, -1):
        if i == n-1:
            vector_x[i] = vector_b[i] / matrix_A[i][i]
        else:
            vector_x[i] = (vector_b[i] - matrix_A[i][i+1]*vector_x[i+1]) / (matrix_A[i][i])
    
    return vector_x



C = cholesky_decomposition(A)
y = forward_substitution(C, vector_b)
CT = C.T
x_A = backsubstitution(CT, y)
print("Rozwiązanie x_A: ", x_A)
```

```python
""" Przykład B """

# definiuje macierz
B = np.zeros((7,7),dtype=float)
np.fill_diagonal(B,4)

for i in range(6):
    B[i][i+1] = 1
    B[i+1][i] = 1
B[6][0] = 1
B[0][6] = 1

B_cholesky = np.linalg.cholesky(B)

"""
B_cholesky * B_cholesky.T * x_B = c

na początku
B_cholesky * d = c

potem: 
B-cholesky.T * x_B = 

"""
vector_c = np.asarray([i for i in range(1,8)])
vector_c = np.float32(vector_c)

vector_d = np.linalg.solve(B_cholesky, vector_c)
                          
x_B = np.linalg.solve(B_cholesky.T, vector_d)
print("Rozwiązanie x_B:", x_B)
```
