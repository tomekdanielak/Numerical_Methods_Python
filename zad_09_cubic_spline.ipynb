```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
%matplotlib inline

vector_x = [-1+(1/32)*i for i in range(65)]
vector_y = [1/(1+5*x**2) for x in vector_x]
plt.scatter(vector_x,vector_y, alpha=0.7, c="red", s=7)

samples = np.linspace(-1,1,1000)

cs = CubicSpline(vector_x,vector_y)
plt.plot(samples, cs(samples))

plt.xlim(-1,1)
plt.show()
```
